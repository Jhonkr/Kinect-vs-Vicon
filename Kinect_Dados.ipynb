```
#Representação gráfica

angulo = df['angulo']
lancamento = df['lancamento']
pega_bola = df['catch'] * 0.5

plt.subplot(2,1,1)
plt.plot(angulo, c = 'm')
plt.title('Kinect Data')
plt.show

plt.subplot(2, 1, 2)
plt.plot(lancamento, c ='r')
plt.plot(pega_bola, c = 'g') 
plt.show()
```
